```python
from datetime import datetime
import pytz

# Define scenarios
scenarios = {
    'summer_morning': datetime(2024, 6, 21, 9, 0),
    'summer_midday': datetime(2024, 6, 21, 12, 0),
    'summer_evening': datetime(2024, 6, 21, 18, 0),
    'winter_morning': datetime(2024, 12, 21, 9, 0),
    'winter_midday': datetime(2024, 12, 21, 12, 0),
    'winter_evening': datetime(2024, 12, 21, 18, 0),
    'spring_midday': datetime(2024, 3, 21, 12, 0),
    'fall_midday': datetime(2024, 9, 21, 12, 0),
}

# Philadelphia location
latitude = 39.9526
longitude = -75.1652
timezone = pytz.timezone('America/New_York')

print("Temporal scenarios defined:")
for name, dt in scenarios.items():
    print(f"  • {name}: {dt.strftime('%B %d, %Y at %I:%M %p')}")

print(f"\nLocation: Philadelphia ({latitude:.4f}°N, {longitude:.4f}°W)")
```

## 3. Calculate Solar Position for Each Scenario

```python
import pvlib

print("Calculating solar position for each scenario...\n")

solar_positions = {}

for scenario_name, dt in scenarios.items():
    # Localize datetime
    dt_local = timezone.localize(dt)
    
    # Calculate solar position
    solar_pos = pvlib.solarposition.get_solarposition(
        dt_local,
        latitude,
        longitude
    )
    
    altitude = solar_pos['apparent_elevation'].values[0]
    azimuth = solar_pos['azimuth'].values[0]
    
    solar_positions[scenario_name] = {
        'altitude': altitude,
        'azimuth': azimuth,
        'datetime': dt
    }
    
    print(f"{scenario_name:20s} - Altitude: {altitude:6.2f}° | Azimuth: {azimuth:6.2f}°")

# Save solar positions
solar_df = pd.DataFrame(solar_positions).T
solar_df.to_csv('data/processed/solar_positions.csv')
print(f"\n✓ Solar positions calculated and saved")
```

## 4. Calculate Building Shadows for Each Scenario

```python
# Project data to PA State Plane (feet) for shadow calculations
CRS_PROJECTED = 'EPSG:2272'

buildings_proj = buildings.to_crs(CRS_PROJECTED)
edges_proj = edges_gdf.to_crs(CRS_PROJECTED)

print(f"Data projected to {CRS_PROJECTED}")
print(f"  Buildings: {len(buildings_proj):,}")
print(f"  Street edges: {len(edges_proj):,}")
```

```python
def calculate_building_shadow(building_geom, height_ft, altitude_deg, azimuth_deg):
    """
    Calculate shadow polygon for a building.
    
    Parameters:
    - building_geom: Building footprint geometry
    - height_ft: Building height in feet
    - altitude_deg: Solar altitude angle in degrees
    - azimuth_deg: Solar azimuth angle in degrees (0=North, 90=East)
    
    Returns:
    - Shadow polygon
    """
    # If sun is below horizon or building has no height, no shadow
    if altitude_deg <= 0 or height_ft <= 0:
        return None
    
    # Calculate shadow length
    altitude_rad = np.radians(altitude_deg)
    shadow_length = height_ft / np.tan(altitude_rad)
    
    # Calculate shadow direction (opposite of sun)
    shadow_azimuth = (azimuth_deg + 180) % 360
    shadow_azimuth_rad = np.radians(shadow_azimuth)
    
    # Calculate shadow offset
    dx = shadow_length * np.sin(shadow_azimuth_rad)
    dy = shadow_length * np.cos(shadow_azimuth_rad)
    
    # Create shadow polygon by translating building footprint
    try:
        from shapely.affinity import translate
        shadow = translate(building_geom, xoff=dx, yoff=dy)
        
        # Union with building footprint for full shadow
        full_shadow = unary_union([building_geom, shadow])
        
        return full_shadow.convex_hull if full_shadow.is_valid else None
    except:
        return None

print("✓ Shadow calculation function defined")
```

```python
print("\nCalculating building shadows for all scenarios...\n")
print("This will take 30-45 minutes for ~16k buildings × 8 scenarios")
print("Please be patient...\n")

# Store shadow geometries for each scenario
building_shadows = {}

for scenario_name, solar_data in solar_positions.items():
    print(f"Processing: {scenario_name}...")
    
    altitude = solar_data['altitude']
    azimuth = solar_data['azimuth']
    
    shadows = []
    
    for idx, building in buildings_proj.iterrows():
        shadow = calculate_building_shadow(
            building.geometry,
            building[height_col],
            altitude,
            azimuth
        )
        
        if shadow is not None:
            shadows.append(shadow)
        
        # Progress indicator
        if (idx + 1) % 2000 == 0:
            print(f"  {idx+1:,} / {len(buildings_proj):,} buildings processed")
    
    # Create GeoDataFrame of shadows
    shadows_gdf = gpd.GeoDataFrame(
        geometry=shadows,
        crs=CRS_PROJECTED
    )
    
    building_shadows[scenario_name] = shadows_gdf
    
    print(f"  ✓ {len(shadows):,} shadows calculated\n")

print("✓ All building shadows calculated")
```

## 5. Extract Tree Canopy Coverage

Using LiDAR heights for shadows

```python
# ============================================================================
# STEP 1: LOAD TREE HEIGHT RASTER
# ============================================================================

import rasterio
from rasterio.mask import mask as raster_mask
from shapely.geometry import box, mapping
from shapely.ops import unary_union
from shapely.affinity import translate
from shapely.strtree import STRtree
import time

print("\n" + "="*70)
print("EFFICIENT TREE SHADOW CALCULATION (SEPARATE + SPATIAL INDEX)")
print("="*70)

tree_height_raster_path = Path('data/processed/tree_heights_from_lidar.tif')

if not tree_height_raster_path.exists():
    print("\n⚠ Tree height raster not found!")
    raise FileNotFoundError("Need tree_heights_from_lidar.tif")

print("\n✓ LiDAR tree HEIGHT raster found")
print("  Strategy: Calculate tree shadows once, use spatial index for querying")
print("  Time estimate: ~3-4 hours total\n")

# Load tree height raster
with rasterio.open(tree_height_raster_path) as src:
    tree_height_data = src.read(1)
    tree_transform = src.transform
    tree_crs = src.crs
    pixel_size = tree_transform[0]

print(f"Tree height raster loaded:")
print(f"  Shape: {tree_height_data.shape}")
print(f"  Mean height: {tree_height_data[tree_height_data > 0].mean():.1f} ft")
print(f"  Max height: {tree_height_data.max():.1f} ft")
print(f"  Pixel size: {pixel_size:.1f} ft")

print("\n✓ Step 1 complete")
```

```python
# ============================================================================
# STEP 2: CREATE SIDEWALK BUFFERS
# ============================================================================

print("\n" + "="*70)
print("CREATING SIDEWALK BUFFERS")
print("="*70)

buffer_distance = 5 * 3.28084  # 5 meters = 16.4 feet per side

print(f"\nBuffer parameters:")
print(f"  Distance per side: {buffer_distance:.1f} feet ({buffer_distance/3.28084:.1f} meters)")
print(f"  Total width: {buffer_distance*2:.1f} feet")

# Ensure correct CRS
if edges_proj.crs.to_epsg() != 2272:
    edges_proj = edges_proj.to_crs('EPSG:2272')

# Create buffers
edges_proj['sidewalk_buffer'] = edges_proj.geometry.buffer(buffer_distance)
edges_proj['buffer_area_sqft'] = edges_proj['sidewalk_buffer'].area

print(f"✓ Created {len(edges_proj):,} sidewalk buffers")
print(f"  Mean area: {edges_proj['buffer_area_sqft'].mean():.0f} sq ft")

print("\n✓ Step 2 complete")
```

```python
# ============================================================================
# STEP 3: CALCULATE TREE SHADOWS (ONLY NEAR PEDESTRIAN NETWORK)
# ============================================================================

print("\n" + "="*70)
print("CALCULATING TREE SHADOWS (NETWORK-FOCUSED)")
print("="*70)

print("\nOptimization: Only processing trees near the pedestrian network")
print("(Ignores trees far from any sidewalk - they can't affect pedestrian shade)\n")

# ============================================================================
# CREATE NETWORK STUDY ZONE (union of all edge buffers)
# ============================================================================

print("Creating network study zone...")

# Calculate maximum possible shadow length (for low sun angle)
max_possible_shadow = 200 / np.tan(np.radians(14))  # Lowest sun ~14° (winter morning)
max_buffer_distance = min(max_possible_shadow, 500)  # Cap at 500ft

print(f"  Maximum shadow length: {max_possible_shadow:.0f} feet")
print(f"  Buffer distance: {max_buffer_distance:.0f} feet")

# Buffer all edges
print(f"  Buffering {len(edges_proj):,} edges...")
edge_study_buffers = edges_proj.geometry.buffer(max_buffer_distance)

# Union all buffers to create study zone
print(f"  Creating union (this takes a few minutes)...")
study_zone_start = time.time()
network_study_zone = unary_union(edge_study_buffers)
study_zone_time = time.time() - study_zone_start

print(f"  ✓ Study zone created in {study_zone_time/60:.1f} minutes")
print(f"  Study zone area: {network_study_zone.area / 5280**2:.2f} square miles")

# ============================================================================
# CALCULATE TREE SHADOWS FOR EACH SCENARIO
# ============================================================================

print(f"\nCalculating tree shadows for {len(scenarios)} scenarios...")
print("Estimated time: 5-8 minutes per scenario (~40-60 min total)")
print("(Much faster - only processing trees near network!)\n")

tree_shadows = {}
tree_shadow_indices = {}

for scenario_name, solar_data in solar_positions.items():
    print(f"{'='*70}")
    print(f"SCENARIO: {scenario_name}")
    print(f"{'='*70}")
    
    scenario_start = time.time()
    
    altitude = solar_data['altitude']
    azimuth = solar_data['azimuth']
    
    # Skip if sun below horizon
    if altitude <= 0:
        print(f"  ⚠ Sun below horizon, skipping\n")
        tree_shadows[scenario_name] = []
        tree_shadow_indices[scenario_name] = None
        continue
    
    # Calculate shadow parameters
    altitude_rad = np.radians(altitude)
    shadow_azimuth = (azimuth + 180) % 360
    shadow_azimuth_rad = np.radians(shadow_azimuth)
    
    print(f"  Sun altitude: {altitude:.1f}° | Shadow direction: {shadow_azimuth:.1f}°")
    
    # ========================================================================
    # ONLY PROCESS TREES WITHIN NETWORK STUDY ZONE
    # ========================================================================
    
    print(f"  Extracting tree pixels within network study zone...")
    
    # Get raster dimensions
    height_pixels, width_pixels = tree_height_data.shape
    
    tree_shadow_geoms = []
    pixels_checked = 0
    pixels_in_zone = 0
    shadow_count = 0
    
    # Process in chunks
    chunk_size = 100
    
    for row_start in range(0, height_pixels, chunk_size):
        row_end = min(row_start + chunk_size, height_pixels)
        
        for row in range(row_start, row_end):
            for col in range(width_pixels):
                tree_height = tree_height_data[row, col]
                
                if tree_height > 0:
                    pixels_checked += 1
                    
                    # Get pixel coordinates
                    px, py = rasterio.transform.xy(tree_transform, row, col)
                    
                    # Create pixel box
                    pixel_box = box(
                        px - pixel_size/2,
                        py - pixel_size/2,
                        px + pixel_size/2,
                        py + pixel_size/2
                    )
                    
                    # ========================================================
                    # CHECK IF PIXEL IS WITHIN NETWORK STUDY ZONE
                    # ========================================================
                    if not network_study_zone.intersects(pixel_box):
                        continue  # Skip this tree - too far from any edge!
                    
                    pixels_in_zone += 1
                    
                    # Calculate shadow for this tree
                    shadow_length = tree_height / np.tan(altitude_rad)
                    dx = shadow_length * np.sin(shadow_azimuth_rad)
                    dy = shadow_length * np.cos(shadow_azimuth_rad)
                    
                    shadow = translate(pixel_box, xoff=dx, yoff=dy)
                    
                    # Create full shadow
                    try:
                        full_shadow = unary_union([pixel_box, shadow])
                        if full_shadow.is_valid:
                            tree_shadow_geoms.append(full_shadow)
                            shadow_count += 1
                    except:
                        pass
            
            # Progress
            if (row + 1) % 500 == 0:
                print(f"    Row {row+1}/{height_pixels} | "
                      f"Checked: {pixels_checked:,} | "
                      f"In zone: {pixels_in_zone:,} | "
                      f"Shadows: {shadow_count:,}")
    
    print(f"\n  ✓ Tree pixels in study area: {pixels_checked:,}")
    print(f"  ✓ Tree pixels near network: {pixels_in_zone:,} ({100*pixels_in_zone/pixels_checked:.1f}%)")
    print(f"  ✓ Created {shadow_count:,} tree shadow polygons")
    print(f"  ✓ Efficiency: Skipped {pixels_checked - pixels_in_zone:,} irrelevant trees!")
    
    # ========================================================================
    # CREATE SPATIAL INDEX
    # ========================================================================
    print(f"  Building spatial index...")
    
    index_start = time.time()
    tree_spatial_index = STRtree(tree_shadow_geoms)
    index_time = time.time() - index_start
    
    print(f"  ✓ Spatial index built in {index_time:.1f} seconds")
    
    # Store shadows and index
    tree_shadows[scenario_name] = tree_shadow_geoms
    tree_shadow_indices[scenario_name] = tree_spatial_index
    
    scenario_time = time.time() - scenario_start
    print(f"  ✓ Scenario complete in {scenario_time/60:.1f} minutes\n")

print("✓ All tree shadows calculated and indexed")

# Calculate total time saved
total_tree_pixels = (tree_height_data > 0).sum()
total_processed = sum([len(tree_shadows[s]) for s in tree_shadows if isinstance(tree_shadows[s], list)])
print(f"\nEfficiency Summary:")
print(f"  Total tree pixels in study area: {total_tree_pixels:,}")
print(f"  Total tree pixels processed: {total_processed:,}")
print(f"  Pixels skipped: {total_tree_pixels - total_processed:,}")
print(f"  Reduction: {100*(total_tree_pixels - total_processed)/total_tree_pixels:.1f}%")

print("\n✓ Step 3 complete")
```

```python
# ============================================================================
# STEP 4: CALCULATE COMBINED SHADE (RASTER-BASED, WITH ERROR HANDLING)
# ============================================================================

import numpy as np
import rasterio
from rasterio.transform import from_origin
from rasterio.features import rasterize
from rasterio.windows import from_bounds
from shapely.geometry import mapping
import time
from pathlib import Path

print("\n" + "="*70)
print("CALCULATING COMBINED SHADE SCORES (RASTER-BASED, MUCH FASTER)")
print("="*70)

# -----------------------------
# 0) SETTINGS
# -----------------------------
OUT_DIR = Path("data/processed")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# Use same resolution you used for tree raster (3 ft)
RES_FT = 3.0

# IMPORTANT: everything in EPSG:2272 (PA State Plane ft)
TARGET_CRS = "EPSG:2272"

# Ensure edges buffers and shadows are projected correctly
if edges_proj.crs is None or str(edges_proj.crs) != TARGET_CRS:
    edges_proj = edges_proj.to_crs(TARGET_CRS)

# We'll rasterize over the full extent of all sidewalk buffers (safe)
minx, miny, maxx, maxy = edges_proj["sidewalk_buffer"].total_bounds

# Add a small padding so edges near boundary don't clip
PAD = 50  # feet
minx -= PAD; miny -= PAD; maxx += PAD; maxy += PAD

width = int(np.ceil((maxx - minx) / RES_FT))
height = int(np.ceil((maxy - miny) / RES_FT))

transform = from_origin(minx, maxy, RES_FT, RES_FT)

print(f"Raster grid:")
print(f"  CRS: {TARGET_CRS}")
print(f"  Resolution: {RES_FT} ft")
print(f"  Size: {width} x {height} = {width*height:,} pixels")

# Helper: compute coverage in a small window quickly
def compute_coverage_for_buffer(buffer_geom, raster_arr, transform, nodata=0):
    """
    Returns fraction of buffer area covered by raster==1.
    Uses a window read from the full raster array + rasterizes the buffer to a mask in that window.
    """
    b = buffer_geom.bounds
    w = from_bounds(*b, transform=transform)

    # Clip window to raster bounds
    row_off = max(0, int(np.floor(w.row_off)))
    col_off = max(0, int(np.floor(w.col_off)))
    row_end = min(raster_arr.shape[0], int(np.ceil(w.row_off + w.height)))
    col_end = min(raster_arr.shape[1], int(np.ceil(w.col_off + w.width)))

    if row_end <= row_off or col_end <= col_off:
        return 0.0

    window_arr = raster_arr[row_off:row_end, col_off:col_end]

    # Build a window-specific transform
    win_transform = rasterio.Affine(
        transform.a, transform.b, transform.c + col_off * transform.a,
        transform.d, transform.e, transform.f + row_off * transform.e
    )

    # Rasterize buffer into this window to create a mask
    mask = rasterize(
        [(mapping(buffer_geom), 1)],
        out_shape=window_arr.shape,
        transform=win_transform,
        fill=0,
        dtype=np.uint8,
        all_touched=True  # keeps behavior closer to pixel-box method
    )

    inside = (mask == 1)
    denom = inside.sum()
    if denom == 0:
        return 0.0

    shaded = (window_arr == 1) & inside
    return shaded.sum() / denom


# -----------------------------
# 1) LOOP SCENARIOS
# -----------------------------
scenario_times = {}

for scenario_name in solar_positions.keys():
    print(f"\n{'='*70}")
    print(f"PROCESSING: {scenario_name}")
    print(f"{'='*70}")
    t0 = time.time()

    # Check if sun is below horizon
    altitude = solar_positions[scenario_name]['altitude']
    
    if altitude <= 0:
        print(f"  ⚠ Sun below horizon (altitude: {altitude:.1f}°), no shadows")
        
        # Create zero arrays for this scenario
        building_cov = np.zeros(len(edges_proj), dtype=float)
        tree_cov = np.zeros(len(edges_proj), dtype=float)
        combined = np.zeros(len(edges_proj), dtype=float)
        
        # Store results
        edges_proj[f"building_shadow_{scenario_name}"] = building_cov
        edges_proj[f"tree_shadow_{scenario_name}"] = tree_cov
        edges_proj[f"shade_{scenario_name}"] = combined
        
        scenario_time = time.time() - t0
        scenario_times[scenario_name] = scenario_time
        
        print(f"  ✓ Skipped (no shadows) in {scenario_time:.2f} seconds\n")
        continue

    # ---- 1A) Rasterize BUILDING shadows for this scenario
    building_shadows_gdf = building_shadows[scenario_name]
    if building_shadows_gdf.crs is None or str(building_shadows_gdf.crs) != TARGET_CRS:
        building_shadows_gdf = building_shadows_gdf.to_crs(TARGET_CRS)

    # Filter valid geometries
    valid_building_geoms = [
        (geom, 1) for geom in building_shadows_gdf.geometry 
        if geom is not None and not geom.is_empty and geom.is_valid
    ]
    
    if len(valid_building_geoms) > 0:
        building_raster = rasterize(
            valid_building_geoms,
            out_shape=(height, width),
            transform=transform,
            fill=0,
            dtype=np.uint8,
            all_touched=True
        )
        print(f"  ✓ Building shadow rasterized ({len(valid_building_geoms):,} geometries)")
    else:
        building_raster = np.zeros((height, width), dtype=np.uint8)
        print("  ⚠ No valid building shadows, using zeros")

    # ---- 1B) Rasterize TREE shadows for this scenario
    tree_geoms = tree_shadows.get(scenario_name, [])
    
    # Filter valid tree geometries
    valid_tree_geoms = [
        (geom, 1) for geom in tree_geoms 
        if geom is not None and not geom.is_empty and geom.is_valid
    ]
    
    if len(valid_tree_geoms) > 0:
        tree_raster = rasterize(
            valid_tree_geoms,
            out_shape=(height, width),
            transform=transform,
            fill=0,
            dtype=np.uint8,
            all_touched=True
        )
        print(f"  ✓ Tree shadow rasterized ({len(valid_tree_geoms):,} geometries)")
    else:
        tree_raster = np.zeros((height, width), dtype=np.uint8)
        print("  ⚠ No valid tree shadows, using zeros")

    # ---- 1C) Compute coverage for each sidewalk buffer (fast numpy)
    building_cov = np.zeros(len(edges_proj), dtype=float)
    tree_cov = np.zeros(len(edges_proj), dtype=float)

    for i, geom in enumerate(edges_proj["sidewalk_buffer"].values):
        if geom is None or geom.is_empty:
            continue

        building_cov[i] = compute_coverage_for_buffer(geom, building_raster, transform)
        tree_cov[i] = compute_coverage_for_buffer(geom, tree_raster, transform)

        if (i + 1) % 2000 == 0:
            elapsed = time.time() - t0
            rate = (i + 1) / elapsed if elapsed > 0 else 0
            eta = (len(edges_proj) - (i + 1)) / rate if rate > 0 else 0
            print(f"    {i+1:,}/{len(edges_proj):,} buffers | ETA ~ {eta/60:.1f} min")

    # Cap at 1.0
    building_cov = np.minimum(building_cov, 1.0)
    tree_cov = np.minimum(tree_cov, 1.0)

    combined = (0.6 * building_cov) + (0.4 * tree_cov)

    # Store back to edges_proj
    edges_proj[f"building_shadow_{scenario_name}"] = building_cov
    edges_proj[f"tree_shadow_{scenario_name}"] = tree_cov
    edges_proj[f"shade_{scenario_name}"] = combined

    scenario_time = time.time() - t0
    scenario_times[scenario_name] = scenario_time

    print(f"\n  ✓ Complete in {scenario_time/60:.1f} minutes")
    print(f"  Building mean: {building_cov.mean():.3f}")
    print(f"  Tree mean:     {tree_cov.mean():.3f}")
    print(f"  Combined mean: {combined.mean():.3f}")
    print(f"  Segments >50%: {(combined > 0.5).sum():,}")

print("\n✓ All shade scores calculated (raster-based)")
print(f"Total time: {sum(scenario_times.values())/60:.1f} minutes")

print("\n✓ Step 4 complete")
```

```python
# ============================================================================
# STEP 4: CALCULATE COMBINED SHADE ON SIDEWALK BUFFERS (RASTER-BASED) - old
# ============================================================================

import numpy as np
import rasterio
from rasterio.transform import from_origin
from rasterio.features import rasterize
from rasterio.windows import from_bounds
from shapely.geometry import mapping
import time
from pathlib import Path

print("\n" + "="*70)
print("CALCULATING COMBINED SHADE SCORES (RASTER-BASED, MUCH FASTER)")
print("="*70)

# -----------------------------
# 0) SETTINGS
# -----------------------------
OUT_DIR = Path("data/processed")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# Use same resolution you used for tree raster (3 ft)
RES_FT = 3.0

# IMPORTANT: everything in EPSG:2272 (PA State Plane ft)
TARGET_CRS = "EPSG:2272"

# Ensure edges buffers and shadows are projected correctly
if edges_proj.crs is None or str(edges_proj.crs) != TARGET_CRS:
    edges_proj = edges_proj.to_crs(TARGET_CRS)

# We'll rasterize over the full extent of all sidewalk buffers (safe)
minx, miny, maxx, maxy = edges_proj["sidewalk_buffer"].total_bounds

# Add a small padding so edges near boundary don't clip
PAD = 50  # feet
minx -= PAD; miny -= PAD; maxx += PAD; maxy += PAD

width = int(np.ceil((maxx - minx) / RES_FT))
height = int(np.ceil((maxy - miny) / RES_FT))

transform = from_origin(minx, maxy, RES_FT, RES_FT)

print(f"Raster grid:")
print(f"  CRS: {TARGET_CRS}")
print(f"  Resolution: {RES_FT} ft")
print(f"  Size: {width} x {height} = {width*height:,} pixels")

# Helper: compute coverage in a small window quickly
def compute_coverage_for_buffer(buffer_geom, raster_arr, transform, nodata=0):
    """
    Returns fraction of buffer area covered by raster==1.
    Uses a window read from the full raster array + rasterizes the buffer to a mask in that window.
    """
    b = buffer_geom.bounds
    w = from_bounds(*b, transform=transform)

    # Clip window to raster bounds
    row_off = max(0, int(np.floor(w.row_off)))
    col_off = max(0, int(np.floor(w.col_off)))
    row_end = min(raster_arr.shape[0], int(np.ceil(w.row_off + w.height)))
    col_end = min(raster_arr.shape[1], int(np.ceil(w.col_off + w.width)))

    if row_end <= row_off or col_end <= col_off:
        return 0.0

    window_arr = raster_arr[row_off:row_end, col_off:col_end]

    # Build a window-specific transform
    win_transform = rasterio.Affine(
        transform.a, transform.b, transform.c + col_off * transform.a,
        transform.d, transform.e, transform.f + row_off * transform.e
    )

    # Rasterize buffer into this window to create a mask
    mask = rasterize(
        [(mapping(buffer_geom), 1)],
        out_shape=window_arr.shape,
        transform=win_transform,
        fill=0,
        dtype=np.uint8,
        all_touched=True  # keeps behavior closer to pixel-box method
    )

    inside = (mask == 1)
    denom = inside.sum()
    if denom == 0:
        return 0.0

    shaded = (window_arr == 1) & inside
    return shaded.sum() / denom


# -----------------------------
# 1) LOOP SCENARIOS
# -----------------------------
scenario_times = {}

for scenario_name in solar_positions.keys():
    print(f"\n{'='*70}")
    print(f"PROCESSING: {scenario_name}")
    print(f"{'='*70}")
    t0 = time.time()

    # ---- 1A) Rasterize BUILDING shadows for this scenario
    building_shadows_gdf = building_shadows[scenario_name]
    if building_shadows_gdf.crs is None or str(building_shadows_gdf.crs) != TARGET_CRS:
        building_shadows_gdf = building_shadows_gdf.to_crs(TARGET_CRS)

    building_raster = rasterize(
        [(geom, 1) for geom in building_shadows_gdf.geometry if geom is not None and not geom.is_empty],
        out_shape=(height, width),
        transform=transform,
        fill=0,
        dtype=np.uint8,
        all_touched=True
    )
    print("  ✓ Building shadow rasterized")

    # ---- 1B) Rasterize TREE shadows for this scenario
    # tree_shadows[scenario_name] is a list of shapely polygons
    tree_geoms = tree_shadows.get(scenario_name, [])
    if tree_geoms and len(tree_geoms) > 0:
        tree_raster = rasterize(
            [(geom, 1) for geom in tree_geoms if geom is not None and not geom.is_empty],
            out_shape=(height, width),
            transform=transform,
            fill=0,
            dtype=np.uint8,
            all_touched=True
        )
        print(f"  ✓ Tree shadow rasterized ({len(tree_geoms):,} polygons)")
    else:
        tree_raster = np.zeros((height, width), dtype=np.uint8)
        print("  ⚠ No tree shadows in this scenario (or missing), using zeros")

    # ---- 1C) Compute coverage for each sidewalk buffer (fast numpy)
    building_cov = np.zeros(len(edges_proj), dtype=float)
    tree_cov = np.zeros(len(edges_proj), dtype=float)

    for i, geom in enumerate(edges_proj["sidewalk_buffer"].values):
        if geom is None or geom.is_empty:
            continue

        building_cov[i] = compute_coverage_for_buffer(geom, building_raster, transform)
        tree_cov[i] = compute_coverage_for_buffer(geom, tree_raster, transform)

        if (i + 1) % 2000 == 0:
            elapsed = time.time() - t0
            rate = (i + 1) / elapsed if elapsed > 0 else 0
            eta = (len(edges_proj) - (i + 1)) / rate if rate > 0 else 0
            print(f"    {i+1:,}/{len(edges_proj):,} buffers | ETA ~ {eta/60:.1f} min")

    # Cap at 1.0
    building_cov = np.minimum(building_cov, 1.0)
    tree_cov = np.minimum(tree_cov, 1.0)

    combined = (0.6 * building_cov) + (0.4 * tree_cov)

    # Store back to edges_proj
    edges_proj[f"building_shadow_{scenario_name}"] = building_cov
    edges_proj[f"tree_shadow_{scenario_name}"] = tree_cov
    edges_proj[f"shade_{scenario_name}"] = combined

    scenario_time = time.time() - t0
    scenario_times[scenario_name] = scenario_time

    print(f"\n  ✓ Complete in {scenario_time/60:.1f} minutes")
    print(f"  Building mean: {building_cov.mean():.3f}")
    print(f"  Tree mean:     {tree_cov.mean():.3f}")
    print(f"  Combined mean: {combined.mean():.3f}")
    print(f"  Segments >50%: {(combined > 0.5).sum():,}")

print("\n✓ All shade scores calculated (raster-based)")
print(f"Total time: {sum(scenario_times.values())/60:.1f} minutes")

print("\n✓ Step 4 complete")
```

```python
# ============================================================================
# EXPORT DATA FOR INTERACTIVE MAP
# Run this AFTER completing Notebook 2 Steps 1-8
# ============================================================================

import json
from pathlib import Path

print("Preparing data for interactive map...")

# Create output directory
map_data_dir = Path('map_data')
map_data_dir.mkdir(exist_ok=True)

# 1. Export nodes (simplified - just unique coordinates)
print("\n1. Extracting nodes...")

all_coords = set()
for idx, row in edges_final.iterrows():
    coords = list(row.geometry.coords)
    all_coords.add(coords[0])
    all_coords.add(coords[-1])

nodes = []
node_id_map = {}
for i, (lon, lat) in enumerate(sorted(all_coords)):
    nodes.append({
        'id': i,
        'lon': lon,
        'lat': lat
    })
    node_id_map[(lon, lat)] = i

print(f"   Extracted {len(nodes):,} nodes")

# 2. Export edges with shade data
print("\n2. Preparing edges...")

edges_data = []
for idx, row in edges_final.iterrows():
    coords = list(row.geometry.coords)
    start_node = node_id_map[coords[0]]
    end_node = node_id_map[coords[-1]]
    
    # Get all shade scores
    shade_scores = {}
    for col in edges_final.columns:
        if col.startswith('shade_') and 'shadow' not in col:
            scenario = col.replace('shade_', '')
            shade_scores[scenario] = float(row[col])
    
    edge_data = {
        'id': idx,
        'source': start_node,
        'target': end_node,
        'length': float(row['length']) if 'length' in row else float(row.geometry.length),
        'geometry': [[lon, lat] for lon, lat in coords],
        'shade': shade_scores,
        'name': row.get('name', '')
    }
    edges_data.append(edge_data)

print(f"   Prepared {len(edges_data):,} edges")

# 3. Export scenario metadata
print("\n3. Creating scenario metadata...")

scenarios = []
for scenario_name, solar_data in solar_positions.items():
    scenarios.append({
        'id': scenario_name,
        'name': scenario_name.replace('_', ' ').title(),
        'altitude': float(solar_data['altitude']),
        'azimuth': float(solar_data['azimuth'])
    })

# 4. Save as JSON files
print("\n4. Saving JSON files...")

with open(map_data_dir / 'nodes.json', 'w') as f:
    json.dump(nodes, f)
print(f"   ✓ nodes.json ({len(nodes):,} nodes)")

with open(map_data_dir / 'edges.json', 'w') as f:
    json.dump(edges_data, f)
print(f"   ✓ edges.json ({len(edges_data):,} edges)")

with open(map_data_dir / 'scenarios.json', 'w') as f:
    json.dump(scenarios, f)
print(f"   ✓ scenarios.json ({len(scenarios)} scenarios)")

# 5. Create metadata
metadata = {
    'bounds': {
        'south': float(edges_final.total_bounds[1]),
        'west': float(edges_final.total_bounds[0]),
        'north': float(edges_final.total_bounds[3]),
        'east': float(edges_final.total_bounds[2])
    },
    'num_nodes': len(nodes),
    'num_edges': len(edges_data),
    'num_scenarios': len(scenarios),
    'default_scenario': 'summer_midday'
}

with open(map_data_dir / 'metadata.json', 'w') as f:
    json.dump(metadata, f, indent=2)
print(f"   ✓ metadata.json")

print("\n✓ Map data export complete!")
print(f"\nFiles saved to: {map_data_dir.absolute()}")
```
