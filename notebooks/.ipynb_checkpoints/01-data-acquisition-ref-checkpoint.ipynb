This notebook acquires all the data needed for shade-optimized routing:
1. OSMnx pedestrian network for University City
2. SEPTA transit stops
3. Tree canopy data
4. Building footprints with heights

It will all be saved so it can be used for next steps in the other notebooks

What This Notebook Does

Sets up project structure (data folders, outputs folders)
Defines study area (University City boundary)
Downloads pedestrian network from OSM using OSMnx
Gets SEPTA transit stops (destinations for routing)
Downloads building footprints (for shadow calculation)
Provides instructions for tree canopy data (manual download needed)
Creates visualization maps for each dataset
Saves all data in data/processed/ for next steps

```python
# PART 1: IMPORTS AND SETUP

# Standard data manipulation (from Week 1-2)
import pandas as pd
import numpy as np

# Geospatial libraries (from Week 3-5)
import geopandas as gpd
from shapely.geometry import Point, Polygon, box

# Network analysis (from Week 7)
import osmnx as ox

# Raster data (from Week 6)
import rasterio
from rasterio.plot import show
from rasterstats import zonal_stats

# Visualization (from Week 3)
import matplotlib.pyplot as plt
import seaborn as sns

# For file management
import os
from pathlib import Path

# Set display options to see more data
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

# Set figure style (like in class)
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("‚úì All libraries imported successfully!")
```

```python
# PART 2: CREATE DATA DIRECTORIES

# Create folders to organize data 

# Create a list of directories we need
directories = [
    'data',           # Raw data
    'data/raw',       # Unprocessed data
    'data/processed', # Cleaned data ready for analysis
    'outputs',        # Maps and figures
    'outputs/maps',
    'outputs/figures'
]

# Loop through and create each directory if it doesn't exist
for directory in directories:
    Path(directory).mkdir(parents=True, exist_ok=True)
    
print("‚úì Directory structure created!")
print("\nProject structure:")
for directory in directories:
    print(f"  üìÅ {directory}/")
```

```python
# ============================================================================
# PART 3: DEFINE STUDY AREA - BOUNDING BOX
# ============================================================================

print("Defining study area: University City, Philadelphia...")

# Create bounding box covering University City
# This includes the main area from ~30th St to ~46th St
minx, miny = -75.230, 39.945  # Southwest corner
maxx, maxy = -75.180, 39.965  # Northeast corner

# Create the bounding box polygon
from shapely.geometry import box
bbox_poly = box(minx, miny, maxx, maxy)

# Convert to GeoDataFrame
study_area = gpd.GeoDataFrame(
    {'name': ['University City Study Area']},
    geometry=[bbox_poly],
    crs='EPSG:4326'
)

print("Study area created!")
print(f"  Bounds: {study_area.total_bounds}")

# Save it
study_area.to_file('data/processed/study_area.geojson', driver='GeoJSON')
print("Saved to: data/processed/study_area.geojson")

# Visualize
fig, ax = plt.subplots(figsize=(10, 10))

study_area.plot(
    ax=ax,
    facecolor='lightblue',
    edgecolor='darkblue',
    linewidth=2,
    alpha=0.3
)

ax.set_title('Study Area: University City', fontsize=16)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('outputs/maps/01_study_area.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n‚úì Study area defined and saved!")
```

```python
# PART 4: GET PEDESTRIAN NETWORK FROM OSM

# We need the walking/pedestrian network (like Week 7)

print("\nDownloading pedestrian network from OpenStreetMap...")

# Configure OSMnx settings (good practice)
ox.settings.use_cache = True  # Cache downloads to speed up repeated runs
ox.settings.log_console = True  # Show progress

# Download the network
# network_type='walk' gets us pedestrian paths, sidewalks, etc.
# simplify=True removes unnecessary nodes (cleaner network)
try:
    G = ox.graph_from_polygon(
        study_area.geometry.iloc[0],  # Get the polygon from our GeoDataFrame
        network_type='walk',           # Pedestrian network
        simplify=True,                 # Remove extra nodes
        retain_all=False               # Only keep the largest connected component
    )
    
    print(f"\n‚úì Network downloaded!")
    print(f"  - Nodes (intersections): {len(G.nodes):,}")
    print(f"  - Edges (street segments): {len(G.edges):,}")
    
    # Save the network (GraphML format - standard for networks)
    ox.save_graphml(G, 'data/processed/university_city_walk_network.graphml')
    print("  - Saved to: data/processed/university_city_walk_network.graphml")
    
except Exception as e:
    print(f"‚ö† Error downloading network: {e}")
    print("  You may need to adjust the study area or try again later.")
    G = None
```

```python
# Quick visualization of the network (like Week 7 lecture)
if G is not None:
    print("\nVisualizing network...")
    
    fig, ax = ox.plot_graph(
        G,
        node_size=5,           # Small dots for intersections
        node_color='red',
        edge_linewidth=0.5,    # Thin lines for streets
        edge_color='gray',
        bgcolor='white',
        show=False,
        close=False
    )
    
    ax.set_title('University City Pedestrian Network', fontsize=16)
    plt.tight_layout()
    plt.savefig('outputs/maps/02_network.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    print("‚úì Network map saved to: outputs/maps/02_network.png")
```

```python
# ============================================================================
# PART 5: CONVERT NETWORK TO GEODATAFRAMES
# ============================================================================

# Convert the network to GeoDataFrames for easier analysis
# This gives us nodes (intersections) and edges (streets) as tables
# (This is a common pattern from Week 7)

if G is not None:
    print("\nConverting network to GeoDataFrames...")
    
    # Get nodes (intersections/endpoints)
    nodes_gdf = ox.graph_to_gdfs(G, edges=False, nodes=True)
    print(f"  - Nodes GeoDataFrame: {len(nodes_gdf)} rows")
    
    # Get edges (street segments)
    edges_gdf = ox.graph_to_gdfs(G, nodes=False, edges=True)
    print(f"  - Edges GeoDataFrame: {len(edges_gdf)} rows")
    
    # IMPORTANT: Reset the index to save properly
    # The MultiIndex from OSMnx causes issues with GeoJSON
    nodes_gdf_reset = nodes_gdf.reset_index()
    edges_gdf_reset = edges_gdf.reset_index()
    
    print("\n  Preparing to save (this may take a moment)...")
    
    # For edges, we need to handle list columns
    # GeoJSON doesn't support list types, so convert them to strings
    for col in edges_gdf_reset.columns:
        # Check if column contains lists
        if edges_gdf_reset[col].apply(lambda x: isinstance(x, list)).any():
            # Convert lists to comma-separated strings
            edges_gdf_reset[col] = edges_gdf_reset[col].apply(
                lambda x: ','.join(map(str, x)) if isinstance(x, list) else x
            )
    
    # Save as GeoJSON
    try:
        nodes_gdf_reset.to_file('data/processed/network_nodes.geojson', 
                                driver='GeoJSON')
        print("  ‚úì Nodes saved")
    except Exception as e:
        print(f"  ‚ö† Could not save nodes: {e}")
    
    try:
        edges_gdf_reset.to_file('data/processed/network_edges.geojson', 
                                driver='GeoJSON')
        print("  ‚úì Edges saved")
    except Exception as e:
        print(f"  ‚ö† Could not save edges: {e}")
    
    print("\n‚úì Network GeoDataFrames saved!")
    
    # Display first few rows to see what we have (like in class)
    print("\nSample of edges data:")
    # Show just the most important columns
    important_cols = ['u', 'v', 'key', 'length', 'highway', 'geometry']
    # Only show columns that exist
    cols_to_show = [col for col in important_cols if col in edges_gdf_reset.columns]
    print(edges_gdf_reset[cols_to_show].head())
    
    # Keep the reset versions for later use
    nodes_gdf = nodes_gdf_reset
    edges_gdf = edges_gdf_reset
```

```python
# ============================================================================
# PART 6: GET OFFICIAL SEPTA TRANSIT STOPS WITH CONNECTIVITY ANALYSIS
# ============================================================================

import json

print("\n" + "="*70)
print("LOADING OFFICIAL SEPTA STOPS DATA")
print("="*70)

# Load the official SEPTA GeoJSON
json_path = 'data/raw/Transit_Stops_(Spring_2025).geojson'

print(f"Loading from: {json_path}")
septa_all = gpd.read_file(json_path)

print(f"‚úì Loaded {len(septa_all):,} SEPTA stops system-wide")

# Filter to University City study area
septa_in_area = gpd.sjoin(septa_all, study_area, how='inner', predicate='intersects')
septa_in_area = septa_in_area.drop_duplicates(subset=['StopId'])
if 'index_right' in septa_in_area.columns:
    septa_in_area = septa_in_area.drop(columns=['index_right'])

print(f"‚úì Found {len(septa_in_area)} stops in University City")

# Categorize by transit type
def categorize_septa_stop(line_abbr):
    line = str(line_abbr).upper()
    rail_lines = ['AIR', 'CHE', 'CHW', 'CYN', 'FOX', 'LAN', 'MED', 'NOR', 
                  'PAO', 'TRE', 'WAR', 'WIL', 'WTR']
    subway_lines = ['MFL', 'BSL', 'BSS', 'NHSL']
    trolley_lines = ['10', '11', '13', '34', '36', '101', '102', 'GRHL']
    
    if line in rail_lines:
        return 'Regional Rail'
    elif line in subway_lines:
        return 'Subway'
    elif line in trolley_lines:
        return 'Trolley'
    else:
        return 'Bus'

septa_in_area['transit_type'] = septa_in_area['LineAbbr'].apply(categorize_septa_stop)

print("\nStops by type:")
for transit_type, count in septa_in_area['transit_type'].value_counts().items():
    print(f"  {transit_type}: {count}")

# Get major transit stops (Subway - Market-Frankford Line)
major_transit = septa_in_area[septa_in_area['transit_type'] == 'Subway'].copy()
major_stations_unique = major_transit.groupby('StopName').first().reset_index()
major_stations_gdf = gpd.GeoDataFrame(
    major_stations_unique,
    geometry=major_stations_unique['geometry'],
    crs=septa_in_area.crs
)

print(f"\n‚úì Major transit stations: {len(major_stations_gdf)}")
for idx, row in major_stations_gdf.iterrows():
    print(f"  ‚Ä¢ {row['StopName']}")

# Add nearest bus stops for connectivity analysis
print("\n" + "-"*70)
print("ADDING CONNECTING BUS STOPS FOR TRANSIT CONNECTIVITY")
print("-"*70)

n_bus_stops = 10  # Number of bus stops per major station

bus_stops = septa_in_area[septa_in_area['transit_type'] == 'Bus'].copy()

# Project for distance calculations
major_proj = major_stations_gdf.to_crs('EPSG:2272')
bus_proj = bus_stops.to_crs('EPSG:2272')

selected_bus_stops = []

for idx, station in major_proj.iterrows():
    bus_proj['distance'] = bus_proj.geometry.distance(station.geometry)
    nearest = bus_proj.nsmallest(n_bus_stops, 'distance').copy()
    nearest['connects_to'] = station['StopName']
    selected_bus_stops.append(nearest)

all_selected_bus = pd.concat(selected_bus_stops, ignore_index=True)
unique_bus_stops = all_selected_bus.drop_duplicates(subset=['StopId']).to_crs('EPSG:4326')

print(f"‚úì Selected {len(unique_bus_stops)} connecting bus stops")

# Combine major transit + bus stops
major_transit['stop_category'] = 'Major Transit'
unique_bus_stops['stop_category'] = 'Feeder Bus'

cols_common = ['StopId', 'StopName', 'LineAbbr', 'transit_type', 'stop_category', 'geometry']
final_stops = pd.concat([
    major_transit[cols_common],
    unique_bus_stops[cols_common]
], ignore_index=True)

# Rename and save
final_stops = final_stops.rename(columns={
    'StopName': 'name',
    'transit_type': 'type',
    'stop_category': 'category'
})

final_stops.to_file('data/processed/septa_stops.geojson', driver='GeoJSON')
septa_gdf = final_stops.copy()

print(f"\n‚úì Final dataset: {len(final_stops)} stops")
print(f"  ‚Ä¢ Major transit: {len(final_stops[final_stops['category'] == 'Major Transit'])}")
print(f"  ‚Ä¢ Connecting bus: {len(final_stops[final_stops['category'] == 'Feeder Bus'])}")
print("‚úì Saved to: data/processed/septa_stops.geojson")

# Visualize
print("\nCreating visualization...")

fig, ax = plt.subplots(figsize=(16, 14))

# Study area and network
study_area.boundary.plot(ax=ax, color='blue', linewidth=3, linestyle='--', 
                         label='Study Area', zorder=1)
edges_gdf.plot(ax=ax, color='lightgray', linewidth=0.3, alpha=0.2, zorder=2)

# Major transit stations
major_plot = final_stops[final_stops['category'] == 'Major Transit']
major_plot.plot(ax=ax, color='darkred', markersize=500, marker='*',
               edgecolor='black', linewidth=2, 
               label=f'Major Transit ({len(major_plot)})', zorder=10)

# Bus stops
bus_plot = final_stops[final_stops['category'] == 'Feeder Bus']
bus_plot.plot(ax=ax, color='green', markersize=80, marker='o',
             edgecolor='darkgreen', linewidth=1,
             label=f'Connecting Bus ({len(bus_plot)})', alpha=0.7, zorder=9)

# Labels for major stations
for idx, row in major_plot.iterrows():
    ax.annotate(row['name'], xy=(row.geometry.x, row.geometry.y),
               xytext=(15, 15), textcoords='offset points',
               fontsize=10, fontweight='bold',
               bbox=dict(boxstyle='round,pad=0.6', facecolor='white',
                        edgecolor='darkred', linewidth=2, alpha=0.95),
               zorder=11)

ax.set_title(f'SEPTA Transit Stops - University City\n' + 
            f'Major Stations + Connecting Bus Stops ({len(final_stops)} total)',
            fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Longitude', fontsize=12)
ax.set_ylabel('Latitude', fontsize=12)
ax.legend(fontsize=12, loc='upper right', framealpha=0.95, edgecolor='black')

plt.tight_layout()
plt.savefig('outputs/maps/03_septa_stops.png', dpi=150, bbox_inches='tight')
plt.show()

print("‚úì Map saved to: outputs/maps/03_septa_stops.png")

print("\n" + "="*70)
print("Transit connectivity dataset ready for shade routing analysis!")
print("="*70)
```

```python
# ============================================================================
# PART 7: GET BUILDING FOOTPRINTS AND HEIGHTS
# ============================================================================

print("\nGetting building footprints...")

# Philadelphia Building Footprints from OpenDataPhilly
buildings_url = "https://hub.arcgis.com/api/v3/datasets/ab9e89e1273f445bb265846c90b38a96_0/downloads/data?format=geojson&spatialRefId=4326&where=1%3D1"

print(f"Downloading from OpenDataPhilly...")
print("(This may take 2-5 minutes - large file...)\n")

try:
    # Read buildings directly from URL (like Week 4)
    print("  Reading buildings data...")
    buildings = gpd.read_file(buildings_url)
    
    print(f"  ‚úì Downloaded {len(buildings):,} building footprints for Philadelphia")
    
    # Check what columns we have
    print(f"\n  Available columns: {list(buildings.columns)[:10]}...")  # Show first 10
    
    # Filter to only buildings in our study area
    print("\n  Filtering to University City study area...")
    
    # Make sure CRS matches
    if buildings.crs != study_area.crs:
        buildings = buildings.to_crs(study_area.crs)
    
    # Spatial filter - keep only buildings that intersect our study area
    buildings_filtered = gpd.sjoin(
        buildings,
        study_area,
        how='inner',
        predicate='intersects'
    )
    
    # Remove duplicate buildings (sjoin can create duplicates)
    buildings_filtered = buildings_filtered.drop_duplicates(
        subset=['geometry']
    )
    
    print(f"  ‚úì {len(buildings_filtered):,} buildings in University City")
    
    # Check for height information
    print("\n  Checking for height data...")
    height_cols = [col for col in buildings_filtered.columns 
                   if 'height' in col.lower() or 'elev' in col.lower()]
    
    if height_cols:
        print(f"  ‚úì Found height columns: {height_cols}")
        
        # Check if we have actual height values
        for col in height_cols:
            non_null = buildings_filtered[col].notna().sum()
            print(f"    - {col}: {non_null} non-null values")
    else:
        print("  ‚ö† No height columns found")
        print("    We'll estimate heights from number of stories if available")
        
        # Check for stories
        story_cols = [col for col in buildings_filtered.columns 
                      if 'story' in col.lower() or 'floor' in col.lower()]
        if story_cols:
            print(f"    Found story columns: {story_cols}")
    
    # Save the filtered buildings
    print("\n  Saving buildings...")
    
    # Drop the index_right column from sjoin if it exists
    if 'index_right' in buildings_filtered.columns:
        buildings_filtered = buildings_filtered.drop(columns=['index_right'])
    
    buildings_filtered.to_file(
        'data/processed/buildings_university_city.geojson',
        driver='GeoJSON'
    )
    print("  ‚úì Saved to: data/processed/buildings_university_city.geojson")
    
    # Quick visualization
    print("\n  Creating building map...")
    fig, ax = plt.subplots(figsize=(12, 12))
    
    # Plot buildings
    buildings_filtered.plot(
        ax=ax,
        facecolor='lightgray',
        edgecolor='darkgray',
        linewidth=0.3,
        alpha=0.7
    )
    
    # Plot SEPTA stops on top
    septa_gdf.plot(
        ax=ax, 
        color='red', 
        markersize=200, 
        marker='*',
        edgecolor='darkred', 
        linewidth=2, 
        zorder=5
    )
    
    # Add station labels
    for idx, row in septa_gdf.iterrows():
        ax.annotate(
            row['name'],
            xy=(row.geometry.x, row.geometry.y),
            xytext=(5, 5),
            textcoords='offset points',
            fontsize=9,
            bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8)
        )
    
    ax.set_title('Building Footprints in University City', fontsize=16)
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    
    plt.tight_layout()
    plt.savefig('outputs/maps/04_buildings.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    print("  ‚úì Building map saved to: outputs/maps/04_buildings.png")
    
except Exception as e:
    print(f"‚ö† Error downloading buildings: {e}")
    print("\nAlternative: Download manually from:")
    print("  https://opendataphilly.org/datasets/building-footprints/")
    buildings_filtered = None
```

```python
# ============================================================================
# PART 8: READ LAND COVER FROM GEODATABASE
# ============================================================================

import fiona
import rasterio
from rasterio.mask import mask
from osgeo import gdal

print("\n" + "="*70)
print("READING PHILADELPHIA LAND COVER FROM GDB")
print("="*70)

# Path to your GDB
gdb_path = 'data/raw/PPR_LandCover_2018.gdb'

print(f"\nGDB location: {gdb_path}")

# Method 1: List layers using fiona
print("\n" + "-"*70)
print("LAYERS IN GEODATABASE:")
print("-"*70)

try:
    layers = fiona.listlayers(gdb_path)
    print(f"Found {len(layers)} layer(s):")
    for i, layer in enumerate(layers, 1):
        print(f"  {i}. {layer}")
    
    # The raster is usually in a layer called something like:
    # 'LandCover', 'PhillyLandCover', 'Philly_LandCover_2018', etc.
    
    # Find the likely raster layer
    raster_layer = None
    for layer in layers:
        if 'land' in layer.lower() or 'cover' in layer.lower():
            raster_layer = layer
            break
    
    if not raster_layer:
        raster_layer = layers[0]  # Use first layer as fallback
    
    print(f"\nUsing layer: '{raster_layer}'")
    
except Exception as e:
    print(f"Could not list layers with fiona: {e}")
    raster_layer = None

# Method 2: Access raster directly using GDAL
print("\n" + "-"*70)
print("ACCESSING RASTER DATA:")
print("-"*70)

# GDB rasters need special handling - try different paths
possible_paths = [
    gdb_path,  # Try the GDB directly
    f"{gdb_path}/LandCover",
    f"{gdb_path}/PhillyLandCover",
    f"{gdb_path}/Philly_LandCover_2018",
    f"{gdb_path}/PPR_LandCover_2018",
]

if raster_layer:
    possible_paths.insert(1, f"{gdb_path}/{raster_layer}")

src = None
for path in possible_paths:
    try:
        print(f"\nTrying: {path}")
        src = rasterio.open(path)
        print(f"‚úì Success! Opened raster from: {path}")
        raster_path = path
        break
    except Exception as e:
        print(f"  ‚úó Failed: {str(e)[:50]}")
        continue

if src is None:
    print("\n‚ö† Could not open raster from GDB using standard methods")
    print("\nLet's try converting the GDB to TIF first...")
    
    # Use GDAL to convert
    print("\nConverting GDB raster to TIF...")
    
    # This command finds and converts the raster
    import subprocess
    
    output_tif = 'data/raw/landcover_philly_full.tif'
    
    # Try with gdalwarp
    try:
        cmd = ['gdalwarp', '-of', 'GTiff', gdb_path, output_tif]
        result = subprocess.run(cmd, capture_output=True, text=True, timeout=300)
        
        if result.returncode == 0:
            print(f"‚úì Converted to: {output_tif}")
            raster_path = output_tif
            src = rasterio.open(raster_path)
        else:
            print(f"gdalwarp failed: {result.stderr}")
            src = None
    except Exception as e:
        print(f"Conversion failed: {e}")
        src = None

# If we have the raster open, examine it
if src:
    print("\n" + "="*70)
    print("RASTER PROPERTIES:")
    print("="*70)
    
    print(f"  Dimensions: {src.width} x {src.height} pixels")
    print(f"  Resolution: {src.res[0]:.2f} x {src.res[1]:.2f} meters")
    print(f"  CRS: {src.crs}")
    print(f"  Bounds: {src.bounds}")
    print(f"  Number of bands: {src.count}")
    
    # Read a small sample to check values
    print("\nReading sample data...")
    window = rasterio.windows.Window(0, 0, 100, 100)
    sample = src.read(1, window=window)
    unique_vals = np.unique(sample)
    
    print(f"  Sample values found: {unique_vals}")
    
    if len(unique_vals) > 10:
        print(f"  (Showing first 10: {unique_vals[:10]})")
    
    # Check if these look like land cover classes (1-7)
    valid_classes = [1, 2, 3, 4, 5, 6, 7]
    has_valid = any(val in valid_classes for val in unique_vals)
    
    if has_valid:
        print("  ‚úì Contains valid land cover classes!")
    else:
        print("  ‚ö† Values don't match expected classes (1-7)")
        
    src.close()
else:
    print("\n‚ö† Could not open raster")
    print("\nPlease check:")
    print(f"1. GDB path is correct: {gdb_path}")
    print("2. GDB contains raster data")
    print("3. GDAL is properly installed")
    raster_path = None
```

```python
# ============================================================================
# PART 8B: CLIP RASTER TO UNIVERSITY CITY STUDY AREA
# ============================================================================

print("\n" + "="*70)
print("CLIPPING RASTER TO UNIVERSITY CITY")
print("="*70)

# The raster path that worked
raster_path = 'data/raw/PPR_LandCover_2018.gdb'

with rasterio.open(raster_path) as src:
    
    print(f"\nFull raster: {src.width} x {src.height} pixels")
    print(f"CRS: {src.crs}")
    
    # Reproject study area to match raster CRS (EPSG:2272)
    print("\nReprojecting study area to match raster...")
    study_area_reproj = study_area.to_crs(src.crs)
    
    print(f"Study area CRS: {study_area.crs} ‚Üí {study_area_reproj.crs}")
    
    # Clip the raster to study area
    print("\nClipping raster to study area...")
    print("(This may take 30-60 seconds for large raster...)")
    
    try:
        out_image, out_transform = mask(
            src,
            study_area_reproj.geometry,
            crop=True,
            all_touched=True,
            filled=True,
            nodata=0
        )
        
        # Update metadata
        out_meta = src.meta.copy()
        out_meta.update({
            "driver": "GTiff",
            "height": out_image.shape[1],
            "width": out_image.shape[2],
            "transform": out_transform,
            "compress": "lzw"  # Compress to save space
        })
        
        # Save clipped raster
        clipped_path = 'data/raw/landcover_university_city.tif'
        
        with rasterio.open(clipped_path, "w", **out_meta) as dest:
            dest.write(out_image)
        
        print(f"\n‚úì Clipped raster saved!")
        print(f"  Output: {clipped_path}")
        print(f"  Size: {out_image.shape[2]} x {out_image.shape[1]} pixels")
        
        # Check what values we have in the clipped area
        print("\nChecking land cover classes in University City...")
        unique_vals = np.unique(out_image)
        print(f"  Unique values: {unique_vals}")
        
        # Count pixels per class
        print("\nPixel counts by class:")
        for val in unique_vals:
            count = np.sum(out_image == val)
            percent = (count / out_image.size) * 100
            
            class_names = {
                0: 'NoData',
                1: 'Tree Canopy',
                2: 'Grass/Shrub',
                3: 'Bare Earth',
                4: 'Water',
                5: 'Buildings',
                6: 'Roads',
                7: 'Other Paved'
            }
            
            class_name = class_names.get(val, f'Unknown ({val})')
            print(f"  {class_name:.<30} {count:>10,} pixels ({percent:>5.1f}%)")
        
        # Check if we have valid classes
        valid_classes = [1, 2, 3, 4, 5, 6, 7]
        has_valid = any(val in valid_classes for val in unique_vals)
        
        if has_valid:
            print("\n‚úì SUCCESS! Valid land cover classes found in University City!")
            raster_path = clipped_path
        else:
            print("\n‚ö† WARNING: No valid land cover classes found")
            print("  This may indicate the study area is outside the raster coverage")
            raster_path = None
            
    except Exception as e:
        print(f"\n‚ö† Error clipping raster: {e}")
        import traceback
        traceback.print_exc()
        raster_path = None
```

```python
# ============================================================================
# PART 8C: EXTRACT TREE CANOPY AND CREATE VISUALIZATIONS
# ============================================================================

print("\n" + "="*70)
print("EXTRACTING TREE CANOPY FROM LAND COVER")
print("="*70)

raster_path = 'data/raw/landcover_university_city.tif'

with rasterio.open(raster_path) as src:
    
    # Read the full land cover
    landcover = src.read(1)
    
    print(f"Loaded land cover: {landcover.shape[1]} x {landcover.shape[0]} pixels")
    
    # Extract tree canopy (class 1)
    tree_canopy_mask = (landcover == 1).astype(np.uint8)
    
    # Calculate statistics
    total_pixels = landcover.size
    tree_pixels = np.sum(tree_canopy_mask)
    tree_percent = (tree_pixels / total_pixels) * 100
    
    # Calculate area (0.5m x 0.5m pixels)
    pixel_area = 0.5 * 0.5  # square meters
    tree_area_sqm = tree_pixels * pixel_area
    tree_area_acres = tree_area_sqm * 0.000247105
    tree_area_sqkm = tree_area_sqm / 1_000_000
    
    print(f"\n‚úì Tree Canopy Statistics:")
    print(f"  Total tree pixels: {tree_pixels:,}")
    print(f"  Coverage: {tree_percent:.1f}%")
    print(f"  Area: {tree_area_sqm:,.0f} sq meters")
    print(f"  Area: {tree_area_acres:.1f} acres")
    print(f"  Area: {tree_area_sqkm:.2f} sq km")
    
    # Save tree canopy as separate raster
    tree_raster_path = 'data/processed/tree_canopy_university_city.tif'
    
    meta = src.meta.copy()
    meta.update({
        'dtype': 'uint8',
        'nodata': 0,
        'compress': 'lzw'
    })
    
    with rasterio.open(tree_raster_path, 'w', **meta) as dst:
        dst.write(tree_canopy_mask, 1)
    
    print(f"\n‚úì Saved tree canopy raster to: {tree_raster_path}")
    
    # Create visualizations
    print("\nCreating visualizations...")
    
    fig, axes = plt.subplots(1, 2, figsize=(18, 8))
    
    # Plot 1: Full land cover with proper colors
    ax1 = axes[0]
    
    colors = {
        0: '#FFFFFF',  # NoData - white
        1: '#228B22',  # Tree canopy - forest green
        2: '#90EE90',  # Grass/shrub - light green
        3: '#D2691E',  # Bare earth - brown
        4: '#4169E1',  # Water - blue
        5: '#8B4513',  # Buildings - saddle brown
        6: '#696969',  # Roads - dim gray
        7: '#A9A9A9'   # Other paved - dark gray
    }
    
    # Create RGB image
    colored = np.zeros((*landcover.shape, 3), dtype=np.uint8)
    for val, color in colors.items():
        mask_val = (landcover == val)
        rgb = tuple(int(color.lstrip('#')[i:i+2], 16) for i in (0, 2, 4))
        colored[mask_val] = rgb
    
    ax1.imshow(colored)
    ax1.set_title('Philadelphia Land Cover - University City\n(2018 High Resolution - 0.5m)', 
                 fontsize=14, fontweight='bold')
    ax1.axis('off')
    
    # Add legend
    from matplotlib.patches import Patch
    legend_elements = [
        Patch(facecolor='#228B22', label=f'Tree Canopy ({tree_percent:.1f}%)'),
        Patch(facecolor='#90EE90', label='Grass/Shrub (14.0%)'),
        Patch(facecolor='#D2691E', label='Bare Earth (1.7%)'),
        Patch(facecolor='#4169E1', label='Water (2.0%)'),
        Patch(facecolor='#8B4513', label='Buildings (27.5%)'),
        Patch(facecolor='#696969', label='Roads (15.9%)'),
        Patch(facecolor='#A9A9A9', label='Other Paved (18.8%)')
    ]
    ax1.legend(handles=legend_elements, loc='upper right', fontsize=10, 
              framealpha=0.9, edgecolor='black')
    
    # Plot 2: Tree canopy only
    ax2 = axes[1]
    
    # Create a nicer tree canopy visualization
    tree_display = np.ma.masked_where(tree_canopy_mask == 0, tree_canopy_mask)
    
    ax2.imshow(np.ones_like(landcover) * 0.95, cmap='gray', vmin=0, vmax=1)  # Light gray background
    ax2.imshow(tree_display, cmap='Greens', alpha=0.9, vmin=0, vmax=1)
    ax2.set_title(f'Tree Canopy Coverage\n{tree_percent:.1f}% ({tree_area_acres:.1f} acres)', 
                 fontsize=14, fontweight='bold')
    ax2.axis('off')
    
    plt.tight_layout()
    plt.savefig('outputs/maps/05_tree_canopy.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    print("‚úì Visualization saved to: outputs/maps/05_tree_canopy.png")
    
    # Create an overlay with SEPTA stops
    print("\nCreating map with SEPTA stops...")
    
    fig, ax = plt.subplots(figsize=(14, 14))
    
    # Show tree canopy
    ax.imshow(tree_display, cmap='Greens', alpha=0.8, vmin=0, vmax=1, 
             extent=[src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top])
    
    # Overlay study area boundary
    study_area_reproj = study_area.to_crs(src.crs)
    study_area_reproj.boundary.plot(ax=ax, color='blue', linewidth=3, 
                                    linestyle='--', label='Study Area')
    
    # Overlay SEPTA stops (reproject to match)
    septa_reproj = septa_gdf.to_crs(src.crs)
    septa_reproj.plot(ax=ax, color='red', markersize=300, marker='*', 
                     edgecolor='darkred', linewidth=2, label='SEPTA Stops', zorder=10)
    
    # Add labels for SEPTA stops
    for idx, row in septa_reproj.iterrows():
        ax.annotate(row['name'], 
                   xy=(row.geometry.x, row.geometry.y),
                   xytext=(10, 10), textcoords='offset points',
                   fontsize=9, fontweight='bold',
                   bbox=dict(boxstyle='round,pad=0.5', facecolor='white', 
                            edgecolor='red', alpha=0.8))
    
    ax.set_title('Tree Canopy Coverage with SEPTA Transit Stops\nUniversity City, Philadelphia', 
                fontsize=16, fontweight='bold', pad=20)
    ax.set_xlabel('Easting (meters)', fontsize=12)
    ax.set_ylabel('Northing (meters)', fontsize=12)
    ax.legend(fontsize=12, loc='upper right')
    
    plt.tight_layout()
    plt.savefig('outputs/maps/06_tree_canopy_with_transit.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    print("‚úì Transit overlay saved to: outputs/maps/06_tree_canopy_with_transit.png")

print("\n" + "="*70)
print("‚úì TREE CANOPY EXTRACTION COMPLETE!")
print("="*70)
```
